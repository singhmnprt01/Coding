```python
class Node:
    def __init__(self, val):
        self.val=val
        self.left=None
        self.right=None
```

```python
root = Node(1)
root.left=Node(3)
root.left.left=Node(5)
root.left.right=Node(6)

root.right=Node(7)
root.right.right=Node(8)
```

```python
root.left.val
```

```python
result = 0
def rootSum(root,path):
    global result 
    
    if not root:
        return
    if not root.left and not root.right:
        result = result + path*10 + root.val  
        return 
    
    rootSum(root.left, path = path*10 + root.val) 
    rootSum(root.right, path = path*10 + root.val) 
    return result
```

```python
rootSum(root, 0)
```
