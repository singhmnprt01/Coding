```python
class Node:
    def __init__(self, item):
        self.left=None
        self.right=None
        self.item=item
 
 
        
#Driver Code to add nodes to trees
root=Node(30)

root.left = Node(40)
root.left.left=Node(50)
root.left.right=Node(70)
root.left.right.left=Node(20)
root.left.right.left=Node(10)

root.right = Node(80)
root.right.left = Node(5)
```

### Naive Solution with O(n^2)

```python
def height(myroot):
    if myroot==None:
        return 0
    else:
        return max(height(myroot.left), height(myroot.right)) + 1
```

```python
# Time complexity is O(n^2)
def checkBalanced(root):
    if root==None:
        return 1
    if root.left is None and root.right is None:
        return 1 
    else:
        leftheight = height(root.left)
        rightheight = height(root.right)
        # print(leftheight, " ", rightheight)
        if abs(leftheight-rightheight) <=1:
            return min(checkBalanced(root.left), checkBalanced(root.right))
        else:
            return 0 
```

```python
checkBalanced(root)
```

### O(n) Solution

```python
def isBalanced(root):
    if root==None:
        return 0 
    leftheight = isBalanced(root.left)
    if leftheight==-1:
        return -1
    
    rightheight = isBalanced(root.height)
    if rightheight==-1:
        return -1
    
    if abs(leftheight=rightheight)>1:
        return -1
    
    return max(leftheight,rightheight) + 1
```

```python
def isBalancedMain(root):
    if isBalanced==-1:
        return False
    else:
        return True
```
