```python
n = 9
a,b,c=2,2,2
```

```python
# Time Complexity: O(3^n)
# Space Complexity: 

def maxSequence(n,a,b,c):
    
    if n==0:
        return 0
    if n <=-1:
        return -1
    max_seq = max(maxSequence(n-a, a,b,c), maxSequence(n-b, a,b,c), maxSequence(n-c,a,b,c))
    
    if max_seq==-1:
        return -1 
    return max_seq + 1
```

```python
maxSequence(n,a,b,c)
```
