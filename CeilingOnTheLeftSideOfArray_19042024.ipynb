```python
import random



x = 1  # lower bound
y = 10000  # upper bound
size = 5000  # size of the array

random_array = random.sample(range(x, y+1), size)
print(random_array)
```

```python
arr=random_array
```

```python
import sys
```

### Simple Solution

```python
%%time
# Time complexity: O(n^2)
# space complexity: O(1)
for i in range(1, len(arr)):
    diff = sys.maxsize
    
    for j in range(0,i):
        if arr[j]>=arr[i]:
            diff = min(diff, arr[j]-arr[i])
    
    if diff==sys.maxsize:
        print("-1", end=" ")
    else:
        print(arr[i]+ diff, end=" ")
```

```python
%%time

# Time complexity is nlog(n) or n*O(n)
s = set()
s.add(arr[0])

for i in range(1,len(arr)):
    it = [x for x in s if x>=arr[i]] ## This can be max O(n)
    
    if len(it)==0:
        print("-1", end=" ")
    else:
        print(min(it), end=" ")
    
    s.add(arr[i])
    
```
